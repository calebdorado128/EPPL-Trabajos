## Formula para calcular area
$A = (B * H) / 2$
---
## Datos de entrada

```python
# Calcular el Area de un triangulo
# Variables B, H, A
# Formula A = (B * H) / 2
# Capturar el valor de B
# Capturar el valor de H
B = float(input("Ingrese el valor de la base: "))
H = float(input("Ingrese el valor de la altura: "))
A = (B * H) / 2
print("El area del triangulo es: ", A)
print("Programa elaborado por Caleb Dorado 0039")
```
